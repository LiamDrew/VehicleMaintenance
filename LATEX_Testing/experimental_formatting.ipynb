```python
from fpdf import FPDF
import random
import string

# Create a PDF document
pdf = FPDF()

# Function to generate random text
def generate_random_text():
    return ''.join(random.choice(string.ascii_uppercase + string.ascii_lowercase) for _ in range(20))

# Add a page to the document
pdf.add_page()

# Create 5 tables
for i in range(5):
    # Generate random text
    random_text = generate_random_text()
    
    # Write the random text
    pdf.set_xy(50, 700 - i*250)
    pdf.set_font("Arial", "B", 16)
    pdf.write(8, random_text)
    
    # Create a table
    pdf.set_xy(50, 500 - i*250)
    pdf.set_font("Arial", "B", 10)
    pdf.set_fill_color(255, 255, 255)  # White background
    pdf.cell(200, 10, txt='Col 1', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Col 2', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Col 3', border=1, ln=True, fill=True)
    pdf.ln()
    pdf.cell(200, 10, txt='Row 1, Col 1', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Row 1, Col 2', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Row 1, Col 3', border=1, ln=True, fill=True)
    pdf.ln()
    pdf.cell(200, 10, txt='Row 2, Col 1', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Row 2, Col 2', border=1, ln=True, fill=True)
    pdf.cell(200, 10, txt='Row 2, Col 3', border=1, ln=True, fill=True)
    pdf.ln()

# Save the PDF
pdf.output("sample.pdf", "F")
```

```python
from fpdf import FPDF
import random
import string

# Create a PDF object
pdf = FPDF()

# Add a page
pdf.add_page()

# Function to generate random text
def generate_random_text():
    return ''.join(random.choice(string.ascii_uppercase + string.ascii_lowercase) for _ in range(20))

# Create 5 tables
for i in range(5):
    # Generate random text
    random_text = generate_random_text()
    
    # Draw the random text
    pdf.set_font("Arial", size=12)
    pdf.cell(10, 10, txt="• ", ln=False, align='L')  # Bullet point
    pdf.cell(180, 10, txt=random_text, ln=True, align='L')  # Text
    
    # Create a table
    table_data = [['Col 1', 'Col 2', 'Col 3'],
                   ['Row 1, Col 1', 'Row 1, Col 2', 'Row 1, Col 3'],
                   ['Row 2, Col 1', 'Row 2, Col 2', 'Row 2, Col 3']]
    
    # Draw the table
    pdf.set_x(20)  # Indent the table to the right of the bullet points
    for row in table_data:
        for data in row:
            pdf.cell(50, 10, data, 1, 0, 'C')
        pdf.ln()
    
    # Add a space after each table
    pdf.ln(10)

# Save the PDF
pdf.output("sample.pdf")
```
